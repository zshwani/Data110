```python
!pip install plotly
import plotly.express as px # Import the plotly.express module and alias it as 'px'

# Group data by latitude, longitude, and gender, then count occurrences
heatmap_data = data.groupby(['Latitude', 'Longitude', 'Gender']).size().reset_index(name='Count')

# Separate data for males and females
male_data = heatmap_data[heatmap_data['Gender'] == 'Male']
female_data = heatmap_data[heatmap_data['Gender'] == 'Female']

# Create male heatmap
fig = px.density_mapbox( # Now 'px' is recognized and the function can be called
    male_data,
    lat='Latitude',
    lon='Longitude',
    z='Count',
    radius=10,  # Adjust as needed
    center=dict(lat=39.2904, lon=-76.6122),  # Example: Baltimore center
    zoom=10,
    mapbox_style='carto-positron',
    title="Interactive Heatmap of Arrests by Gender"
)
fig.update_traces(name="Male Arrests", visible=True)

# Create female heatmap
fig2 = px.density_mapbox(
    female_data,
    lat='Latitude',
    lon='Longitude',
    z='Count',
    radius=10,
    center=dict(lat=39.2904, lon=-76.6122),
    zoom=10,
    mapbox_style='carto-positron'
)
fig2.update_traces(name="Female Arrests", visible='legendonly')

# Combine layers
fig.add_traces(fig2.data)

# Show the combined heatmap
fig.show()
```

```python
map_center = [39.2904, -76.6122]  # Example: Baltimore center
m = folium.Map(location=map_center, zoom_start=10)
heat_map=cleaned_df[['Latitude','Longitude']].values
HeatMap(heat_map,radius=15).add_to(m)
map_path="combined_heatmap.html"
m.save(map_path)
map_path
```
